```python
file_path = '/Users/gawain/Desktop/6.Real-time Local Business Condition Index/RTLBCI-2023-04-28.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def convert_df(sub_df):
      # transpose the dataframe
      sub_df_T = sub_df.T
      # rename the first row to 'Value'
      sub_df_T = sub_df_T.rename(columns={1: 'Value'})
      # remove the first row
      sub_df_T = sub_df_T.iloc[1:]
      # add the name to the first column
      sub_df_T = sub_df_T.rename_axis('Date').reset_index()
      # convert the date to datetime
      sub_df_T = sub_df_T.astype({'Date': 'datetime64[ns]'})
      # convert the value to float
      sub_df_T = sub_df_T.astype({'Value': 'float'})
      return sub_df_T
```

```python
# Take halifax as an example
sub_df = df.iloc[1:2]
co_df = convert_df(sub_df)
co_df
```

```python
co_df.info()
```

```python
plt.plot(co_df['Date'], co_df['Value'])
```
